```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import pickle
import time
```

```python
pkl_file = open('train_set.pkl', 'rb')
train_set = pickle.load(pkl_file)
pkl_file.close()

pkl_file = open('test_set.pkl', 'rb')
test_set = pickle.load(pkl_file)
pkl_file.close()

pkl_file = open('val_set.pkl', 'rb')
val_set = pickle.load(pkl_file)
pkl_file.close()
```

```python
train_set[train_set['year'] == 1994].plot(column='ratio_best', figsize=(15, 15), vmin = 0, vmax = train_set['ratio_best'].max())
plt.show()
```

```python
test_set.plot(column='ratio_best', figsize=(15, 15), vmin = 0, vmax = train_set['ratio_best'].max())
plt.show()
```

```python
val_set.plot(column='ratio_best', figsize=(15, 15), vmin = 0, vmax = train_set['ratio_best'].max())
plt.show()
```

```python
def corretion(val_set, test_set, train_set):

    start = time.time()

    for gid in val_set['gid'].unique():
        
        ratio_best = train_set[train_set['gid'] == gid]['ratio_best'].mean()
        
        val_set.loc[(val_set['gid'] == gid), 'ratio_best'] = ratio_best
        test_set.loc[(test_set['gid'] == gid), 'ratio_best'] = ratio_best

    end = time.time()

    runtime = (end - start)/60
    print(runtime)
```

```python
corretion(val_set, test_set, train_set)
```
